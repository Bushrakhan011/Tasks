```python
import numpy as np
x = np.array([1,2,3,4,5])
print("The array is : ",x)
```

```python
x = np.ones(10)
print("The array of 1s  is : ",x)
```

```python
x = np.zeros(5)
print("The array of 0s is : ",x)
```

```python
x = np.arange(1,10)
print(x)
x = np.arange(5,11)
print(x)
x = np.arange(1,10,2)   # arrange from 1 to 10 with 2 steps
print(x)
```

```python
x = np.linspace(1,10)
print(x)
print( len(x) )   # total no of elements in x
```

```python
x = x**2    # taking square of all values of x
print(x)
```

```python
x = np.sqrt(x)    # taking sqraue root of the above squared x
print(x)   
```

```python
x1 = x+x   # adding the array x with itself
print(x1)
```

```python
x2 = x*x
print(x2)
```

```python
x3 = x/x
print(x3)
```
